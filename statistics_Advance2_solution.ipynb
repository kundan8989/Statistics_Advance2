### 1. What is hypothesis testing in statistics 
##### Hypothesis testing in statistics is a method used to make decisions or inferences about population parameters based on sample data. It helps you test assumptions (called hypotheses) about a population.

Hypothesis:
A hypothesis is a claim or assumption about a population parameter


### 2.What is the null hypothesis, and how does it differ from the alternative hypothesis
#### What is the Null Hypothesis?
###### The null hypothesis, is a default assumption or claim that there is no effect, no difference, or no relationship between variables.
###### It acts as the starting point in hypothesis testing.
###### It is what we try to disprove or reject with evidence from sample data.


#### What is the Alternative Hypothesis?
###### It is a contrary claim to the null hypothesis. It states that there is an effect, a difference, or a relationship.
###### It is what the researcher wants to prove or support.


### 3.What is the significance level in hypothesis testing, and why is it important

### 4. What Does a P-value Represent in Hypothesis Testing?
The p-value is the probability of obtaining a test result at least as extreme as the one observed, assuming the null hypothesis  is true.

In Simple Terms:
The p-value tells you how likely your sample result is due to random chance.

##### A small p-value ‚Üí observed data is unlikely under ùêª0 ‚Üí evidence against H0
##### A large p-value ‚Üí observed data is likely under ùêª0‚Üí not enough evidence to reject ùêª0



### 5.How to Interpret the P-value in Hypothesis Testing
The p-value helps you decide whether to reject the null hypothesis (ùêª0) based on how extreme your observed data is assuming ùêª0 is true.

 #### Key Interpretation:
The p-value is the probability of observing a test statistic as extreme as the one from your data, if the null hypothesis is true.

### 6.What are Type 1 and Type 2 errors in hypothesis testing
##### Type I and Type II errors are fundamental concepts in hypothesis testing that describe incorrect conclusions about the null hypothesis. A Type I error (false positive) occurs when the null hypothesis is rejected despite being true, while a Type II error (false negative) happens when the null hypothesis is not rejected even though it is false. These errors are inversely related, and balancing their risks is critical in statistical analysis.

### 7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing8
#### One-Tailed Test
###### Purpose: Tests for an effect in a specific direction (either greater than or less than a certain value).

###### Alternative Hypothesis: Directional (e.g., mean is greater than or less than a specified value).

###### Critical Region: Entire significance level (e.g., 5%) is placed in one tail of the distribution‚Äîeither left or right.

###### Use Case: When you are only interested in deviations in one direction (e.g., testing if a new drug is more effective, not just different).

###### Statistical Power: Generally higher for detecting an effect in the specified direction compared to a two-tailed test.

#### Two-Tailed Test
###### Purpose: Tests for an effect in both directions (either greater than or less than a certain value).

###### Alternative Hypothesis: Non-directional (e.g., mean is not equal to a specified value).

###### Critical Region: Significance level is split equally between both tails of the distribution (e.g., 2.5% in each tail for a 5% significance level).

###### Use Case: When you want to detect any difference, regardless of direction (e.g., testing if a drug has any effect, positive or negative).

###### Statistical Power: Lower than a one-tailed test for a specific direction, but allows detection of effects in both directions

### 8.What is the Z-test, and when is it used in hypothesis testing
#####  Z-test is a statistical hypothesis test used to determine whether there is a significant difference between sample and population means, or between the means of two populations, under the assumption that the data are approximately normally distributed and the population variance (or standard deviation) is known.

##### When Is a Z-Test Used?
###### Sample Size: The sample size should be large (typically n‚â•30), which, according to the central limit theorem, makes the sampling distribution of the mean approximately normal.

###### Population Standard Deviation: The population standard deviation (or variance) must be known. If it is unknown and the sample size is small, a t-test is generally used instead.

###### Normality: The underlying population should be normally distributed, or the sample size should be large enough for the central limit theorem to apply.

###### Independence: Data points in the sample must be independent of each other

### 10.What is the T-distribution, and when should it be used instead of the normal distribution
What is the T-distribution, and when should it be used instead of the normal distribution8
The t-distribution (or Student‚Äôs t-distribution) is a probability distribution that is symmetric and bell-shaped, similar to the normal distribution, but with heavier tails. This means it gives more probability to values far from the mean, reflecting greater uncertainty when estimating population parameters from small samples.

When to Use the T-Distribution
Use the t-distribution instead of the normal distribution in hypothesis testing when:

#### The sample size is small (typically n<30).
#### The population standard deviation is unknown and must be estimated from the sample.

#### The data are approximately normally distributed, or the sample is sufficiently random.

#### In these situations, the t-distribution provides more accurate critical values and confidence intervals because it accounts for the extra uncertainty from estimating the standard deviation from the sample.

#### As the sample size increases, the t-distribution approaches the normal distribution, and for large samples (typically 
n>30), the difference becomes negligible.



### 11.What is the difference between a Z-test and a T-test
###### Z-test is used when the sample size is large and the population variance (standard deviation) is known. It relies on the standard normal distribution to calculate probabilities and critical values.

###### T-test is used when the sample size is small or the population variance is unknown. It uses the t-distribution, which has heavier tails to account for the extra uncertainty from estimating the standard deviation from the sample

### 12.What is the T-test, and how is it used in hypothesis testing
###### A t-test is a statistical hypothesis test used to determine whether there is a significant difference between the means of two groups. It is particularly useful when the sample size is small and the population standard deviation is unknown.

###### How the T-test Is Used in Hypothesis Testing
###### Purpose: The t-test helps assess whether observed differences between sample means are likely due to random chance or represent a real effect in the population.

##### Process:
Formulate the null hypothesis (typically, that the means of the two groups are equal) and the alternative hypothesis (that the means are different).

###### Calculate the t-statistic using the sample data, which incorporates the difference between means, the sample standard deviations, and the sample sizes.

###### Compare the calculated t-statistic to a critical value from the t-distribution (which depends on the degrees of freedom and desired significance level).

###### Decide whether to reject the null hypothesis based on whether the t-statistic falls in the critical region.

###### Interpretation: If the p-value is less than the chosen significance level (often 0.05), the null hypothesis is rejected, indicating a statistically significant difference between the group means

###### Relationship

#### 14.What is a confidence interval, and how is it used to interpret statistical results
A confidence interval is a statistical range, calculated from sample data, that is likely to contain the true value of an unknown population parameter‚Äîsuch as a mean or proportion‚Äîwithin a specified level of confidence (commonly 95% or 99%). Instead of providing a single point estimate, a confidence interval gives an upper and lower bound, reflecting the uncertainty inherent in using a sample to estimate a population value.

##### Interpreting Statistical Results with Confidence Intervals
Degree of Uncertainty: The width of the interval reflects the degree of uncertainty around the estimate‚Äîa wider interval indicates more uncertainty, while a narrower interval suggests greater precision.

Statistical Significance: In hypothesis testing, if a confidence interval for a difference between groups does not include zero (or the null value), the result is considered statistically significant at the corresponding confidence level.

Practical Use: Confidence intervals help researchers and analysts understand the range within which the true value likely falls, rather than relying solely on a single estimate.

"A 95% confidence interval does not mean there is a 95% probability that the true value is within the interval; rather, it means that 95% of such intervals from repeated samples would contain the true value

### 15.What is the margin of error, and how does it affect the confidence interval
##### What is the Margin of Error?
The margin of error quantifies the amount of uncertainty in a confidence interval. It represents the maximum expected difference between the true population parameter and a sample estimate, given a specific confidence level. Mathematically, it is calculated as:

Margin of Error=Critical Value√óStandard Error

The critical value depends on the chosen confidence level (e.g., 1.96 for 95% confidence).

The standard error reflects the variability in the sample and is typically calculated as the standard deviation divided by the square root of the sample size.

How Does Margin of Error Affect the Confidence Interval?
The confidence interval is constructed as:

#### Sample Estimate¬±Margin of Error

This means the margin of error determines how far above and below the sample estimate the true population parameter is likely to fall.

Width of Confidence Interval: The margin of error is half the width of the confidence interval. A larger margin of error results in a wider confidence interval, indicating more uncertainty. Conversely, a smaller margin of error produces a narrower interval, reflecting greater precision.

#### Factors Influencing Margin of Error:

Sample Size: Larger samples reduce the margin of error, making the confidence interval narrower.

Confidence Level: Higher confidence levels (e.g., 99% vs. 95%) increase the margin of error, resulting in a wider interval, because they require more certainty that the true value is captured.

Variability: Greater variability in the data increases the standard error, and thus the margin of error.

### 16.How is Bayes' Theorem used in statistics, and what is its significance
> Bayes' Theorem is a fundamental principle in statistics that describes how to update the probability of a hypothesis or event based on new evidence or data. It provides a mathematical formula for calculating conditional probability‚Äîthe likelihood of an event occurring given that another event has already occurred.

#### How Bayes' Theorem Is Used in Statistics
Bayesian Inference: One of the primary applications is in Bayesian inference, where Bayes' Theorem is used to update prior beliefs (prior probabilities) about a parameter or hypothesis as new data (evidence) becomes available. The result is a posterior probability, which reflects the updated belief after considering the evidence.

Medical Testing: In medicine, Bayes' Theorem helps interpret diagnostic test results by considering both the accuracy of the test and the prevalence of the condition in the population. This avoids common errors like the base-rate fallacy, where the rarity of a disease is ignored when interpreting positive test results.

Decision Making: It is widely used in fields such as finance, law, engineering, and machine learning to revise risk assessments, forecast probabilities, and make better decisions as new information becomes available.

#### Significance of Bayes' Theorem
Incorporates Prior Knowledge: Unlike classical (frequentist) approaches, Bayes' Theorem allows statisticians to formally incorporate existing knowledge or beliefs (priors) into the analysis, making the results more flexible and context-aware.

Dynamic Updating: It enables continuous updating of probabilities as new evidence is observed, making it especially powerful in real-world situations where information accumulates over time.

Wide Applications: Beyond statistics, Bayes' Theorem underpins methods in machine learning (such as Naive Bayes classifiers), artificial intelligence, medical diagnostics, scientific research, and more

### 17.What is the Chi-square distribution, and when is it used
>The chi-square distribution is a continuous probability distribution that arises from the sum of the squares of independent standard normal random variables. It is characterized by its degrees of freedom, which correspond to the number of independent standard normal variables being squared and summed. The shape of the chi-square distribution depends on these degrees of freedom, becoming more symmetric as the number increases.

#### When Is the Chi-Square Distribution Used?
The chi-square distribution is widely used in inferential statistics, particularly in hypothesis testing and estimation. Its main applications include:

##### Chi-square tests for categorical data:

Goodness-of-fit test: Determines whether observed categorical data fit a specific distribution or expected frequencies.

Test of independence: Assesses whether two categorical variables are independent in a contingency table.

Estimation of variance: Used in constructing confidence intervals and hypothesis tests for population variance when the underlying data are normally distributed.

Model assessment: Measures how well theoretical models fit observed data, especially in fields like genetics, marketing, and survey analysis.

The chi-square test is only applicable to categorical data and is fundamental for analyzing relationships and distributions in such datasets

### 18.What is the Chi-square goodness of fit test, and how is it applied
> What is the Chi-square goodness of fit test, and how is it applied8
The Chi-square goodness of fit test is a statistical hypothesis test used to determine whether the observed distribution of a categorical variable matches an expected distribution based on a specific theoretical model or prior knowledge. It is most commonly applied to categorical (not continuous) data and helps assess whether sample data are representative of a full population or conform to a hypothesized distribution.

#### How the Chi-square Goodness of Fit Test Is Applied
1. Formulate Hypotheses
2. 
Null hypothesis (H‚ÇÄ): The observed frequencies fit the expected distribution (no significant difference).

Alternative hypothesis (H‚ÇÅ): The observed frequencies do not fit the expected distribution (significant difference).

2. Collect Data

Gather counts for each category from a random sample of the population.

3. Calculate Expected Frequencies

Determine the expected frequency for each category based on the hypothesized distribution. For example, if you expect equal proportions across five categories in 100 observations, each category‚Äôs expected frequency is 20.

4. Compute the Chi-square Statistic


5. Determine Degrees of Freedom

For the goodness of fit test, degrees of freedom = (number of categories ‚Äì 1).

6. Find the Critical Value and Compare

Use a chi-square distribution table to find the critical value for your chosen significance level (e.g., 0.05) and degrees of freedom.

7. Make a Decision

If the calculated chi-square statistic is greater than the critical value, reject the null hypothesis (the observed distribution significantly differs from the expected).

If it is less, do not reject the null hypothesis (no significant difference).

##### Example Applications

### 19.What is the F-distribution, and when is it used in hypothesis testing
>What is the F-distribution, and when is it used in hypothesis testing8
The F-distribution is a continuous, right-skewed probability distribution that arises from the ratio of two independent chi-square distributions, each divided by their respective degrees of freedom. It is defined by two parameters: the numerator degrees of freedom (df1) and the denominator degrees of freedom (df2), which shape the distribution. The F-distribution is always non-negative and becomes more symmetric as the degrees of freedom increase.

When Is the F-Distribution Used in Hypothesis Testing?
The F-distribution is primarily used in hypothesis testing to:

Compare variances: It tests whether two population variances are equal, which is fundamental in many statistical procedures.

Analysis of Variance (ANOVA): The most common application is in ANOVA, where the F-test determines if the means of two or more groups are significantly different by comparing the variance between groups to the variance within groups.

Regression analysis: It is used to test the overall significance of a regression model, assessing whether at least one predictor variable has a significant effect

### 20.What is an ANOVA test, and what are its assumptions
> An ANOVA test (Analysis of Variance) is a statistical method used to determine whether there are statistically significant differences between the means of three or more independent groups. Instead of comparing means two at a time (which increases the risk of Type I errors), ANOVA evaluates all groups simultaneously by analyzing the variance both between and within groups.

What Is ANOVA?
Purpose: To test if at least one group mean is different from the others, indicating a real effect rather than random variation.

How It Works: ANOVA partitions the total variance in the data into variance due to differences between group means (‚Äúbetween-group variance‚Äù) and variance within the groups (‚Äúwithin-group variance‚Äù). It then uses the F-test to compare these variances. A large F-statistic suggests that group means are not all equal.


#### Assumptions of ANOVA
To ensure valid results, ANOVA relies on several key assumptions:

Independence: Observations are independent within and across groups.

Normality: The data in each group are approximately normally distributed.

Homogeneity of variances: The variances among the groups are equal (homoscedasticity).

Violating these assumptions can affect the reliability of the ANOVA results.


### 21.What are the different types of ANOVA tests
There are several types of ANOVA (Analysis of Variance) tests, each designed for specific experimental designs and research questions. The main types include:

#### 1.One-Way ANOVA
Purpose: Compares the means of three or more independent groups based on a single factor (independent variable).

Example: Comparing average test scores among students taught with different teaching methods.

#### 2.Two-Way ANOVA (Factorial ANOVA)
Purpose: Examines the effect of two independent variables (factors) on a dependent variable, including possible interaction effects between the factors.

Example: Studying the impact of both teaching method and gender on student performance.

#### 3.Factorial ANOVA
Purpose: An extension of two-way ANOVA, involving more than two factors, allowing for the study of multiple independent variables and their interactions.



### 22.What is the F-test, and how does it relate to hypothesis testing?
>The F-test is a statistical test that uses the F-distribution to compare the variances of two populations or samples, or to assess the overall significance of a statistical model. In hypothesis testing, the F-test is most commonly applied in two scenarios:

Testing Equality of Variances: The F-test evaluates whether the variances of two independent samples are significantly different. The test statistic is calculated as the ratio of the two sample variances. If this ratio (the F-statistic) is significantly larger or smaller than expected under the null hypothesis (that the variances are equal), the null hypothesis is rejected.

Analysis of Variance (ANOVA): The F-test is the foundation of ANOVA, where it determines if the variability between group means is significantly greater than the variability within groups. A significant F-statistic indicates that at least one group mean is different from the others.

#### How the F-test Relates to Hypothesis Testing
Null Hypothesis (H‚ÇÄ): Typically states that the variances are equal, or that all group means are equal (in ANOVA).

Alternative Hypothesis (H‚ÇÅ): States that at least one variance is different, or at least one group mean is different.

Decision Rule: The calculated F-statistic is compared to a critical value from the F-distribution table (based on degrees of freedom and significance level). If the F-statistic exceeds the critical value, the null hypothesis is rejected

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)

import numpy as np
import matplotlib.pyplot as plt

# Define range of discrete uniform distribution
low = 1     # Start value (inclusive)
high = 7    # End value (exclusive in numpy, so this gives 1 to 6)

# Possible values
x = np.arange(low, high)

# PMF for discrete uniform: Equal probability for each value
pmf = [1 / (high - low)] * len(x)

# Plotting the PMF
plt.stem(x, pmf, use_line_collection=True)
plt.title("PMF of Discrete Uniform Distribution (1 to 6)")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.ylim(0, max(pmf) + 0.05)
plt.grid(True)
plt.show()
```

```python
#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")
    
    return p**x * (1 - p)**(1 - x)

# Probability of success
p = 0.7

# Calculate P(X = 1) and P(X = 0)
print("P(X = 1):", bernoulli_pdf(1, p))
print("P(X = 0):", bernoulli_pdf(0, p))


```

```python
# 9. Write a Python function to calculate Z-scores from a dataset and plot them>
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    z_scores = (data - mean) / std_dev
    return z_scores

# Example dataset
data = np.random.normal(50, 10, 100)  # mean=50, std=10, 100 samples

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Plot Z-scores
plt.figure(figsize=(10, 5))
plt.plot(z_scores, marker='o', linestyle='-', color='teal')
plt.axhline(y=0, color='gray', linestyle='--')
plt.axhline(y=1, color='red', linestyle='--', label='Z = 1')
plt.axhline(y=-1, color='red', linestyle='--', label='Z = -1')
plt.title("Z-scores of Dataset")
plt.xlabel("Index")
plt.ylabel("Z-score")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 15. Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

def z_test(sample, pop_mean, pop_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)
    
    # Standard error of the mean
    se = pop_std / np.sqrt(n)
    
    # Z statistic
    z_stat = (sample_mean - pop_mean) / se
    
    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))
    
    # Decision
    reject_null = p_value < alpha
    
    return z_stat, p_value, reject_null

# Example data
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=30)  # Sample drawn from population with mean=52
population_mean = 50     # Known population mean under H0
population_std = 10      # Known population std deviation

# Run Z-test
z_stat, p_val, reject = z_test(sample_data, population_mean, population_std)

print(f"Z statistic: {z_stat:.3f}")
print(f"P-value: {p_val:.4f}")
print("Reject null hypothesis" if reject else "Fail to reject null hypothesis")
```

```python
# 1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
#interpret the results
import numpy as np
from scipy.stats import norm

def one_sample_z_test(sample_data, population_mean, population_std, alpha=0.05):
    # Sample statistics
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    
    # Z-test statistic
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
    
    # Two-tailed test p-value
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Interpretation
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
    else:
        print("Conclusion: Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")

# Example usage
np.random.seed(42)
sample = np.random.normal(loc=52, scale=10, size=30)  # Simulated sample
population_mean = 50
population_std = 10

one_sample_z_test(sample, population_mean, population_std)
```

```python
# 4.Perform a two-tailed Z-test using Python and visualize the decision region on a plot


import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def two_tailed_z_test_with_plot(sample, population_mean, population_std, alpha=0.05):
    # Sample statistics
    sample_mean = np.mean(sample)
    n = len(sample)
    z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
    p_value = 2 * (1 - norm.cdf(abs(z_score)))

    # Output
    print(f"Sample Mean: {sample_mean:.2f}")
    print(f"Z-score: {z_score:.4f}")
    print(f"P-value: {p_value:.4f}")
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis (significant difference).")
    else:
        print("Conclusion: Fail to reject the null hypothesis (no significant difference).")

    # Visualization
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    z_critical = norm.ppf(1 - alpha / 2)

    plt.figure(figsize=(10, 5))
    plt.plot(x, y, label='Standard Normal Distribution')
    
    # Shade rejection regions
    plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region (Left)')
    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3, label='Rejection Region (Right)')
    
    # Plot the Z-score
    plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f'Z-score = {z_score:.2f}')
    
    plt.title('Two-Tailed Z-Test with Decision Regions')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# üîπ Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=52, scale=10, size=30)
population_mean = 50
population_std = 10

two_tailed_z_test_with_plot(sample_data, population_mean, population_std)
```

```python
# 5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testingE

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=50, mu1=53, sigma=10, n=30, alpha=0.05):
    # Step 1: Calculate critical value for Type I error (alpha)
    z_critical = norm.ppf(1 - alpha)
    se = sigma / np.sqrt(n)
    x_critical = mu0 + z_critical * se  # Right-tail test

    # Step 2: Compute Type II error (beta)
    z_beta = (x_critical - mu1) / se
    beta = norm.cdf(z_beta)

    # Step 3: Plot the distributions
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = norm.pdf(x, loc=mu0, scale=se)
    y1 = norm.pdf(x, loc=mu1, scale=se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='Null Hypothesis (H‚ÇÄ)', color='blue')
    plt.plot(x, y1, label='Alternative Hypothesis (H‚ÇÅ)', color='green')

    # Type I Error Region
    plt.fill_between(x, y0, where=(x >= x_critical), color='red', alpha=0.4, label='Type I Error (Œ±)')
    
    # Type II Error Region
    plt.fill_between(x, y1, where=(x < x_critical), color='orange', alpha=0.4, label='Type II Error (Œ≤)')

    # Mark critical value
    plt.axvline(x_critical, color='black', linestyle='--', label=f'Critical Value ‚âà {x_critical:.2f}')

    plt.title('Type I and Type II Errors Visualization')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print error probabilities
    print(f"Critical Value (X): {x_critical:.2f}")
    print(f"Type I Error (Œ±): {alpha:.2f}")
    print(f"Type II Error (Œ≤): {beta:.4f}")
    print(f"Power of the Test (1 - Œ≤): {1 - beta:.4f}")

# üîπ Example usage
visualize_type1_type2_errors(mu0=50, mu1=53, sigma=10, n=30, alpha=0.05)
```

```python
# 6. Write a Python program to perform an independent T-test and interpret the results

import numpy as np
from scipy.stats import ttest_ind

def independent_t_test(group1, group2, alpha=0.05):
    # Perform the independent two-sample t-test
    t_stat, p_value = ttest_ind(group1, group2, equal_var=False)  # Welch's t-test

    # Display results
    print(f"Group 1 Mean: {np.mean(group1):.2f}")
    print(f"Group 2 Mean: {np.mean(group2):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")

    # Interpretation
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis ‚Äî significant difference between groups.")
    else:
        print("Conclusion: Fail to reject the null hypothesis ‚Äî no significant difference between groups.")

# üîπ Example usage: Simulate two groups
np.random.seed(42)
group1 = np.random.normal(loc=50, scale=10, size=30)
group2 = np.random.normal(loc=55, scale=10, size=30)

independent_t_test(group1, group2)
```

```python
# 7.Perform a paired sample T-test using Python and visualize the comparison resultsE

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import ttest_rel

def paired_t_test_visualization(before, after, alpha=0.05):
    # Paired t-test
    t_stat, p_value = ttest_rel(before, after)

    # Output
    print(f"Mean (Before): {np.mean(before):.2f}")
    print(f"Mean (After): {np.mean(after):.2f}")
    print(f"T-statistic: {t_stat:.4f}")
    print(f"P-value: {p_value:.4f}")
    
    if p_value < alpha:
        print("Conclusion: Reject the null hypothesis ‚Äî significant change between paired samples.")
    else:
        print("Conclusion: Fail to reject the null hypothesis ‚Äî no significant change between paired samples.")

    # Visualization - line plot
    plt.figure(figsize=(12, 5))

    # 1. Line plot for paired comparison
    plt.subplot(1, 2, 1)
    for i in range(len(before)):
        plt.plot([0, 1], [before[i], after[i]], marker='o', color='gray')
    plt.xticks([0, 1], ['Before', 'After'])
    plt.ylabel('Scores')
    plt.title('Paired Sample Comparison')
    plt.grid(True)

    # 2. Histogram of differences
    plt.subplot(1, 2, 2)
    differences = after - before
    plt.hist(differences, bins=10, color='skyblue', edgecolor='black')
    plt.axvline(np.mean(differences), color='red', linestyle='--', label='Mean Difference')
    plt.title('Distribution of Differences (After - Before)')
    plt.xlabel('Difference')
    plt.ylabel('Frequency')
    plt.legend()
    plt.tight_layout()
    plt.show()

# üîπ Example usage
np.random.seed(42)
before_scores = np.random.normal(loc=70, scale=5, size=30)
after_scores = before_scores + np.random.normal(loc=2, scale=5, size=30)  # Slight improvement

paired_t_test_visualization(before_scores, after_scores)
```

```python
# 8. Simulate data and perform both Z-test and T-test, then compare the results using PythonE

import numpy as np
from scipy.stats import ttest_1samp, norm
import matplotlib.pyplot as plt

def z_test(sample, pop_mean, pop_std):
    """Perform one-sample Z-test."""
    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)
    z = (sample_mean - pop_mean) / se
    p = 2 * (1 - norm.cdf(abs(z)))  # Two-tailed
    return z, p

def compare_z_t_tests():
    # Simulate a sample dataset
    np.random.seed(0)
    population_mean = 100
    population_std = 15
    sample = np.random.normal(loc=102, scale=15, size=30)  # Simulated sample

    # Perform Z-test (known population std)
    z_stat, z_p = z_test(sample, pop_mean=population_mean, pop_std=population_std)

    # Perform T-test (unknown population std)
    t_stat, t_p = ttest_1samp(sample, population_mean)

    # Display results
    print("Z-test (assumes known population std):")
    print(f"  Z-statistic: {z_stat:.4f}")
    print(f"  P-value: {z_p:.4f}")

    print("\nT-test (uses sample std):")
    print(f"  T-statistic: {t_stat:.4f}")
    print(f"  P-value: {t_p:.4f}")

    # Visualization
    labels = ['Z-test', 'T-test']
    p_values = [z_p, t_p]

    plt.bar(labels, p_values, color=['blue', 'green'])
    plt.axhline(y=0.05, color='red', linestyle='--', label='Œ± = 0.05')
    plt.title("Comparison of P-values: Z-test vs T-test")
    plt.ylabel("P-value")
    plt.legend()
    plt.show()

# Run the comparison
compare_z_t_tests()
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significanceE\

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = t.ppf((1 + confidence) / 2, df=n-1)
    margin = t_crit * std_err
    lower = mean - margin
    upper = mean + margin
    return mean, lower, upper

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=10, size=30)

mean, lower, upper = confidence_interval(sample_data)
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
# 10.Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy.stats import t

def margin_of_error(data, confidence=0.95):
    n = len(data)
    sample_std = np.std(data, ddof=1)  # sample standard deviation
    standard_error = sample_std / np.sqrt(n)
    t_score = t.ppf((1 + confidence) / 2, df=n-1)
    moe = t_score * standard_error
    return moe

# üîπ Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=100, scale=15, size=25)

moe = margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): ¬±{moe:.2f}")
```

```python
# 11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

def bayes_theorem(prior_H, likelihood_E_given_H, prior_not_H, likelihood_E_given_not_H):
    numerator = likelihood_E_given_H * prior_H
    denominator = numerator + likelihood_E_given_not_H * prior_not_H
    posterior_H = numerator / denominator
    return posterior_H

# Given values
prior_H = 0.01              # P(H): disease prevalence
likelihood_E_given_H = 0.99 # P(E|H): test sensitivity
prior_not_H = 1 - prior_H    # P(~H)
likelihood_E_given_not_H = 0.05  # P(E|~H): false positive rate

posterior = bayes_theorem(prior_H, likelihood_E_given_H, prior_not_H, likelihood_E_given_not_H)

print(f"Posterior probability P(H|E) = {posterior:.4f}")
print(f"This means there is a {posterior*100:.2f}% chance the person has the disease given a positive test.")
```

```python
# 12.Perform a Chi-square test for independence between two categorical variables in Python
import pandas as pd
from scipy.stats import chi2_contingency

# Example data: Two categorical variables ‚Äî Gender and Preference
data = {'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
        'Preference': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'Yes', 'No', 'No', 'Yes']}

df = pd.DataFrame(data)

# Create a contingency table
contingency_table = pd.crosstab(df['Gender'], df['Preference'])
print("Contingency Table:")
print(contingency_table)

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))

# Interpretation
alpha = 0.05
if p < alpha:
    print("\nReject the null hypothesis ‚Äî variables are likely associated.")
else:
    print("\nFail to reject the null hypothesis ‚Äî no evidence variables are associated.")
```

```python
# 13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    observed = np.array(observed)
    
    # Calculate row sums, column sums, and total sum
    row_sums = observed.sum(axis=1).reshape(-1, 1)
    col_sums = observed.sum(axis=0).reshape(1, -1)
    total = observed.sum()

    # Calculate expected frequencies formula: (row_sum * col_sum) / total
    expected = (row_sums @ col_sums) / total

    # Convert back to DataFrame if input was DataFrame
    if isinstance(observed, pd.DataFrame):
        expected = pd.DataFrame(expected, index=observed.index, columns=observed.columns)
    return expected

# Example observed data (contingency table)
observed_data = pd.DataFrame({
    'Yes': [20, 30],
    'No': [15, 35]
}, index=['Male', 'Female'])

print("Observed Frequencies:")
print(observed_data)

expected_freq = calculate_expected_frequencies(observed_data)

print("\nExpected Frequencies:")
print(expected_freq)
```

```python
# 14.Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy.stats import chisquare

observed = np.array([50, 30, 20])

# Define expected probabilities (must sum to 1)
expected_probs = np.array([1/3, 1/3, 1/3])

# Convert expected probabilities to expected counts matching observed total
expected = expected_probs * observed.sum()

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: observed data does NOT fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: observed data fits the expected distribution.")
```
